Crime has always been one of the larger problems in India, especially since it is the second most populous country in the world. Also, some of the cultural peculiarities - sex being a taboo in India, the skewed sex ratio, age-old customs like Sati and Dowry and the lower status women hold in society - cause women to be at higher risk than usual. There is also lack of awareness, inadequate infrastructure and insufficient programs in place for their safety and protection. The problem is most severe in rural areas, but some urban areas like Tier 1 cities, and surprisingly, even the capital, are recognized as high risk zones. In recent times, this has become a matter of great concern after a series of very brutal crimes against young girls have been reported and have made international headlines over the past few years. Safety of women has suddenly become a national priority and is even playing an influential role in the upcoming central elections. These motivations compel us to analyze crime data from India with the goal of gaining an insight into the nature and depth of the problem. We use data from the Open Government Data (OGD) Platform India (http://data.gov.in)

Preliminary Analysis To Do

We attempt to calculate the total number of people (men and women) arrested for crime against women in 2012 by summing up the Grand Total column. Further, we try to find the total number of these criminals per thousand. Turns out, about 1 in every thousand people got arrested for commiting a crime against women. This is a big number for a country with a population of over a billion.



```python
currentYear = currentYear[currentYear["Crime head"] != "TOTAL CRIMES AGAINST WOMEN"]
currentYear = currentYear[(currentYear['STATE/UT'] != "TOTAL (STATES)") & (currentYear['STATE/UT'] != "TOTAL (UTs)") & (currentYear['STATE/UT'] != "TOTAL (ALL-INDIA)")]
```

```python
totalCrime_2012 = sum(currentYear['Grand Total'])
print totalCrime_2012

totalIndiaPopulation = 1237000000

criminalsPer_1000 = (totalCrime_2012 *1000) / totalIndiaPopulation 
print criminalsPer_1000
```

```python
currentFile1 = codecs.open(os.path.join(FILES_DIR,'CasesPersonsHeldUnderCAW2010-2012.csv'), encoding='iso-8859-1')
Year2010_2012 = pd.read_csv(currentFile1)
Year2010_2012.head()
```

```python
#Removing extra rows and columns containing totals
Year2010_2012 = Year2010_2012[Year2010_2012["CRIME HEAD"] != "TOTAL CRIMES AGAINST WOMEN"]
Year2010_2012 = Year2010_2012[(Year2010_2012['STATE/UT'] != "TOTAL (STATES)") & (Year2010_2012['STATE/UT'] != "TOTAL (UTs)") & (Year2010_2012['STATE/UT'] != "TOTAL (ALL-INDIA)")]
Year2010_2012.head()
```

```python
#Keeping only columns for Cases
newdf = Year2010_2012[['STATE/UT', 'CRIME HEAD', 'CASES REGISTERED - 2010', 'CASES CHARGESHEETED - 2010', 'CASES CONVICTED - 2010', 'CASES REGISTERED - 2011', 'CASES CHARGESHEETED - 2011', 'CASES CONVICTED - 2011', 'CASES REGISTERED - 2012', 'CASES CHARGESHEETED - 2012', 'CASES CONVICTED - 2012']]
newdf.head()
```

```python
newdf.groupby("CRIME HEAD").sum().plot(kind='barh', title="Cases by Category", figsize=(15,25))
plt.legend(loc='right')
```

```python
#Keeping only columns for Persons
newdf1 = Year2010_2012[['STATE/UT', 'CRIME HEAD', 'PERSONS ARRESTED - 2010', 'PERSONS CHARGESHEETED - 2010', 'PERSONS CONVICTED - 2010', 'PERSONS ARRESTED - 2011', 'PERSONS CHARGESHEETED - 2011', 'PERSONS CONVICTED - 2011', 'PERSONS ARRESTED - 2012', 'PERSONS CHARGESHEETED - 2012', 'PERSONS CONVICTED - 2012']]
newdf1.head()
```

```python
newdf1.groupby("CRIME HEAD").sum().plot(kind='barh', title="Criminals by Category", figsize=(15,25))
plt.legend(loc='right')
```

```python
#Calculating numbers for people acquitted, to get a general idea of how many people get off for these crimes
newdf1['PERSONS ACQUITTED - 2010'] = newdf1['PERSONS ARRESTED - 2010'] - newdf1['PERSONS CONVICTED - 2010']
newdf1['PERSONS ACQUITTED - 2011'] = newdf1['PERSONS ARRESTED - 2011'] - newdf1['PERSONS CONVICTED - 2011']
newdf1['PERSONS ACQUITTED - 2012'] = newdf1['PERSONS ARRESTED - 2012'] - newdf1['PERSONS CONVICTED - 2012']
newdf2 = newdf1[['PERSONS ACQUITTED - 2010', 'PERSONS ACQUITTED - 2011', 'PERSONS ACQUITTED - 2012']]
newdf2.head()
```

```python
#Concentrating on current year, series in descending order of persons acquitted
newlist = newdf2["PERSONS ACQUITTED - 2012"].order(ascending = False)
newlist
```

```python
print(max(newlist))
print(min(newlist))
```

```python
#Mapping the numbers to respective State and Crime Category to create a profile of where most acquittals happen and for what crime
newdf1[['STATE/UT', 'CRIME HEAD', 'PERSONS ACQUITTED - 2012']].sort_index(by="PERSONS ACQUITTED - 2012", ascending=False).head()
```

```python
newDF=newdf1.groupby("CRIME HEAD").sum()
newDF=newDF.sort_index(by="PERSONS ACQUITTED - 2012", ascending=False)
newDF[['PERSONS ACQUITTED - 2012']]
```

```python
newDF1=newDF[['PERSONS ACQUITTED - 2012']]
newDF1.plot(kind='barh', title="Crimes by State", figsize=(10,5))
```

```python
newDF['Percentage Acquitted - 2012'] = (newDF['PERSONS ACQUITTED - 2012']/newDF['PERSONS ARRESTED - 2012'])*100
newDF.head()
```

```python
newDF[['Percentage Acquitted - 2012']]
```

```python
newDF[['Percentage Acquitted - 2012']].sort_index(by = 'Percentage Acquitted - 2012', ascending='False').plot(kind='barh', title="Crimes by State", figsize=(10,8))
```
