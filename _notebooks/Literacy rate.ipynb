```python
import numpy as np
import os
import pandas as pd
import codecs
import matplotlib.pyplot as plt

from pandas import DataFrame, Series
```

```python
FILES_DIR = os.path.join(os.pardir, "_data")
currentFile = codecs.open(os.path.join(FILES_DIR,'literacy_rate_2011.csv'), encoding='iso-8859-1')

# Create the data frame
data = DataFrame(pd.read_csv(currentFile))
data
```

```python
plt.figure
grouped= data.groupby('State').sum().sort_index(by="Literacy Rate", ascending=False)
grouped.plot(kind='bar', figsize=(5,5), title = 'Literacy rate', color='red')
```
