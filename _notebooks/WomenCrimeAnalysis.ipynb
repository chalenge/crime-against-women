```python
import numpy as np
import os
import pandas as pd
import codecs
import matplotlib.pyplot as plt

from pandas import DataFrame, Series
from numpy.random import randn
from IPython.display import display, Image, HTML, clear_output
from IPython.html import widgets
```

We start by importing the data from the csv data files. We then create data frames that we will use to analyze data for the most recent year of data collection for 2012. 

```python
# Set the path to get the files to be the base directory of the application
FILES_DIR = os.path.join(os.pardir, "_data")
# file_path = os.path.join(FILES_DIR, 'CurrentYear2012Data.csv')

# Open the file with current year statistics and create a data frame
currentFile = codecs.open(os.path.join(FILES_DIR,'CurrentYear2012Data.csv'), encoding='iso-8859-1')
currentYear = DataFrame(pd.read_csv(currentFile))

# Remove summary (total) rows from the data
currentYear = currentYear[(currentYear['STATE/UT'] != 'TOTAL (UTs)') & (currentYear['STATE/UT'] != 'TOTAL (STATES)') & (currentYear['STATE/UT'] != 'TOTAL (ALL-INDIA)')]
currentYear = currentYear[(currentYear['Crime head'] != 'TOTAL CRIMES AGAINST WOMEN')]

# Open the file with current year population and create a data frame
currentFile = codecs.open(os.path.join(FILES_DIR,'indiaStatesPopulation.csv'), encoding='iso-8859-1')
currentPopulation = DataFrame(pd.read_csv(currentFile))

# Merge the two dataframes to create a dataframe with current year statistics and population.
currentYear = pd.merge(currentYear, currentPopulation, on='STATE/UT')

#Open the file with historic crime type
currentFile = codecs.open(os.path.join(FILES_DIR,'historicCrimeTypeYearly.csv'), encoding='iso-8859-1')
historicCrimes = DataFrame(pd.read_csv(currentFile))
historicCrimes = historicCrimes[(historicCrimes['CRIME HEAD'] != 'Total Crimes against Women')]
```

CURRENT YEAR ANALYSIS

Our initial analysis, focuses on the current crime numbers for the year 2012. We are interested to find the total number of crimes committed against women. 

```python
print sum(currentYear['Grand Total'])
```

Now we are interested to find the frequency of the current crimes bu each state in India. 

```python
stateNumbers = currentYear.groupby('STATE/UT').sum()
stateNumbers = stateNumbers.sort(['Grand Total'], ascending=[0])
stateNumbers.head()
```

Preliminary Analysis To Do

We attempt to calculate the total number of people (men and women) arrested for crime against women in 2012 by summing up the Grand Total column. Further, we try to find the total number of these criminals per thousand. Turns out, about 0.03 in every thousand people got arrested for commiting a crime against women. This is a big number for a country with a population of over a billion.  

```python
stateNumbers[['Grand Total', 'Total population']] = stateNumbers[['Grand Total', 'Total population']].astype(float)


# stateNumbers['weightedCrime'] = stateNumbers.apply(lambda return x: Grand Total,axis=1)
stateNumbers['Weighted Crime'] = stateNumbers.apply(lambda s: (s['Grand Total'] *100000) / s['Total population'], axis=1)
stateNumbers

totalCrime_2012 = sum(stateNumbers['Grand Total'])
totalPopulation_2012 = sum(stateNumbers['Total population'])

criminalsPer_1000 = (totalCrime_2012 / totalPopulation_2012) * 1000
print criminalsPer_1000
```

After examining the frequency of total crime in each state of India for 2012, we are interested to see if there is a trend for certain crimes over time from 2001 to 2012. We find that the crime of Cruelty by Husband and Relatives is the most frequent crime overall. 

```python
#Plotting the data on axis 1
fig,ax1 = plt.subplots(figsize=(15,10))
ax1.set_title("Crime Information",fontsize=16)
xaxis_years= ['2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012']
p1 = ax1.plot(xaxis_years, historicCrimes.iloc[0][1:], color='b',label="Rape",linewidth=2.0)
p2 = ax1.plot(xaxis_years, historicCrimes.iloc[1][1:], color='purple',label="Kidnapping and Abduction of Women & Girls",linewidth=2.0)
p3 = ax1.plot(xaxis_years, historicCrimes.iloc[2][1:], color='red',label="Dowry Deaths",linewidth=2.0)
p4 = ax1.plot(xaxis_years, historicCrimes.iloc[3][1:], color='black',label="Assault on women with intent to outrage her more",linewidth=2.0)
p5 = ax1.plot(xaxis_years, historicCrimes.iloc[4][1:], color='yellow',label="Insult to the modesty of Women",linewidth=2.0)
p6 = ax1.plot(xaxis_years, historicCrimes.iloc[5][1:], color='pink',label="Cruelty by Husband or his relatives",linewidth=2.0)
p7 = ax1.plot(xaxis_years, historicCrimes.iloc[6][1:], color='green',label="Importation of Girls from Foreign Country",linewidth=2.0)
p8 = ax1.plot(xaxis_years, historicCrimes.iloc[7][1:], color='orange',label="Immoral Traffic (P) Act",linewidth=2.0)
p9 = ax1.plot(xaxis_years, historicCrimes.iloc[8][1:], color='maroon',label="Dowry Prohibition Act",linewidth=2.0)
p10 = ax1.plot(xaxis_years, historicCrimes.iloc[9][1:], color='indigo',label="Indecent Representation of Women (P) Act",linewidth=2.0)
p11 = ax1.plot(xaxis_years, historicCrimes.iloc[10][1:], color='brown',label="Commission of Sati (P) Act",linewidth=2.0)
ax1.set_ylabel('Number of Crimes')
ax1.set_xlabel('Year')

#Creating a list of labels
p = p1+p2+p3+p4+p5+p6+p7+p8+p9+p10+p11
labs = [l.get_label() for l in p]

#Adding legend at the specified position
ax1.legend(p, labs, loc='best')
```
